#JSON

```python
# prompt: load JSON file name data_solution from the drive

import json
import pandas as pd
from google.colab import drive

drive.mount('/content/drive')

# Replace 'data_solution.json' with the actual file name
file_path = '/content/drive/MyDrive/data_solution.json'

try:
    with open(file_path, 'r') as file:
        data = json.load(file)
        # Now you can work with the data loaded from the JSON file
        # For example, if it's a list of dictionaries, you could create a DataFrame:
        if isinstance(data, list) and all(isinstance(item, dict) for item in data):
            df = pd.DataFrame(data)
            print(df.head())  # Display the first few rows of the DataFrame
        else:
            print("JSON data is not in a suitable format to be used as DataFrame.")
            print(data) # print the raw JSON data
except FileNotFoundError:
    print(f"Error: File not found at {file_path}")
except json.JSONDecodeError:
    print(f"Error: Invalid JSON format in file {file_path}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
